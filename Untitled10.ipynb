```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#WARRIORS ANTES E DEPOIS DE STEPHEN CURRY
DataWarriors = pd.read_csv('WarriorsBefore.csv')
DataWarriorsCurry = pd.read_csv('WarriorsCurry.csv')

```

```python
DerWarriorsBef = DataWarriors['L'].sum()
VitWarriorsBef = DataWarriors['W'].sum()
DerWarriorsAft = DataWarriorsCurry['L'].sum()
VitWarriorsAft = DataWarriorsCurry['W'].sum()
WarriorsAntes = [DerWarriorsBef,VitWarriorsBef]
WarriorsDepois = [DerWarriorsAft,VitWarriorsAft]
```

```python
labels = ['Derrotas','Vitorias']
WarriorsAntes = [DerWarriorsBef, VitWarriorsBef]
WarriorsDepois = [DerWarriorsAft, VitWarriorsAft]
x = np.arange(len(labels))  # posições no eixo x
largura = 0.35

fig, ax = plt.subplots()
# barras antes
ax.bar(x - largura/2, WarriorsAntes, largura, label='1992-2008', color=['red','red'])
# barras depois
ax.bar(x + largura/2, WarriorsDepois, largura, label='Depois do Curry', color=['green','green'])

# detalhes do gráfico
ax.set_ylabel('Quantidade')
ax.set_title('Vitórias x Derrotas Warriors')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```
