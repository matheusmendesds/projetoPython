Análise Jogadores NBA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import mplcursors
```

```python
DataBase = pd.read_csv('TabelaNBA.csv')
```

```python
DataBase.rename(
    columns={
        "PTS\n▼":"PTS per Game"
    }, inplace=True)
```

```python
DataBase.columns
```

```python
plt.figure(figsize=(12,6))

plt.bar(DataBase['Player'], DataBase['PTS per Game'])
plt.title('Maiores pontuadores por jogo temporada 2024/2025')
plt.xlabel('Jogadores')
plt.ylabel('Pontos')
plt.xticks(rotation=45, ha="right")

plt.show()
```

```python
Db_Ast = DataBase.sort_values('AST', ascending=False)
```

```python
plt.figure(figsize=(12,6))

plt.bar(Db_Ast['Player'],Db_Ast['AST'])
plt.title('Numeros Totais de Assitências')
plt.xlabel('Jogadores')
plt.ylabel('Assistências')
plt.xticks(rotation=45, ha="right")

plt.show();
```

```python
plt.figure(figsize=(14,6))
sns.scatterplot(data=DataBase, x='Player', y='3P')
plt.title('Numero de Cestas de 3 Pontos por Jogo')
plt.xticks(rotation=45, ha="right")
plt.show();
```

```python
Db_Shots = DataBase[['Player','3P','3PA']]
```

```python
plt.figure(figsize=(16,8))

plt.bar(Db_Shots['Player'], Db_Shots['3P'], label='Arremessos Certos')
plt.bar(Db_Shots['Player'], Db_Shots['3PA'], bottom=Db_Shots['3P'], label='Arremessos Tentandos')
plt.title('Arremessos totais de 3 Pontos')      
plt.ylabel('Arremessos de 3 pontos')
plt.xlabel('Jogadores')
plt.xticks(rotation=45, ha="right")
plt.legend()

plt.show()           
```
