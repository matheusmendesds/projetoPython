Análise Jogadores NBA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
DataBase = pd.read_csv('TabelaNBA.csv')
```

```python
DataBase.rename(
    columns={
        "PTS\n▼":"PTS per Game"
    }, inplace=True)
```

```python
DataBase.columns
```

```python
plt.figure(figsize=(12,6))

plt.bar(DataBase['Player'], DataBase['PTS per Game'])
plt.title('Maiores pontuadores por jogo temporada 2024/2025')
plt.xlabel('Jogadores')
plt.ylabel('Pontos')
plt.xticks(rotation=45, ha="right")

plt.show()
```

```python
Db_Ast = DataBase.sort_values('AST', ascending=False)
```

```python
Db_Ast
```

```python
plt.figure(figsize=(12,6))

plt.bar(Db_Ast['Player'],Db_Ast['AST'].sort_values())
plt.title('Numeros Totais de Assitências')
plt.xlabel('Jogadores')
plt.ylabel('Assistências')
plt.xticks(rotation=45, ha="right")

plt.show();
```
